##Data wrangling

```python
df1[['Unfurnished']]=df1['furnish_type']=='Unfurnished'
df1['Unfurnished']=df1['Unfurnished'].replace(to_replace=[True,False],value=[1,0])
df1[["furnished"]]=df1['furnish_type']=='Furnished'
df1['furnished']=df1['furnished'].replace(to_replace=[True,False],value=[1,0])
df1[["Semi-furnished"]]=df1['furnish_type']=='Semi-Furnished'
df1['Semi-furnished']=df1['Semi-furnished'].replace(to_replace=[True,False],value=[1,0])
print(df1.info())
df1
```

```python
df1[["Owner"]]=df1['seller_type']=='OWNER'
df1['Owner']=df1['Owner'].replace(to_replace=[True,False],value=[1,0])
df1[['Agent']]=df1['seller_type']=='AGENT'
df1['Agent']=df1['Agent'].replace(to_replace=[True,False],value=[1,0])
print(df1.info())
df1
```

```python
df1[['1 bathroom']]=df1['bathroom']=='1 bathrooms'
df1['1 bathroom']=df1['1 bathroom'].replace(to_replace=[True,False],value=[1,0])
df1[['2 bathroom']]=df1['bathroom']=='2 bathrooms'
df1['2 bathroom']=df1['2 bathroom'].replace(to_replace=[True,False],value=[1,0])
df1[['3 bathroom']]=df1['bathroom']=='3 bathrooms'
df1['3 bathroom']=df1['3 bathroom'].replace(to_replace=[True,False],value=[1,0])
df1[['4 bathroom']]=df1['bathroom']=='4 bathrooms'
df1['4 bathroom']=df1['4 bathroom'].replace(to_replace=[True,False],value=[1,0])
df1[['5 bathroom']]=df1['bathroom']=='5 bathrooms'
df1['5 bathroom']=df1['5 bathroom'].replace(to_replace=[True,False],value=[1,0])
df1[['6 bathroom']]=df1['bathroom']=='6 bathrooms'
df1['6 bathroom']=df1['6 bathroom'].replace(to_replace=[True,False],value=[1,0])
df1[['8 bathroom']]=df1['bathroom']=='8 bathrooms'
df1['8 bathroom']=df1['8 bathroom'].replace(to_replace=[True,False],value=[1,0])
print(df1.info())
df1
```

```python
df1=df1.drop(['bathroom'],axis=1)
df1[['bathroom']]=df1['1 bathroom']+2*df1['2 bathroom']+3*df1['3 bathroom']+4*df1['4 bathroom']+5*df1['5 bathroom']+6*df1['6 bathroom']+8*df1['8 bathroom']
print(df1.info())
df1
```

```python
df1[['Independent Floor']]=df1['property_type']=='Independent Floor'
df1['Independent Floor']=df1['Independent Floor'].replace(to_replace=[True,False],value=[1,0])
df1[['Apartment']]=df1['property_type']=='Apartment'
df1['Apartment']=df1['Apartment'].replace(to_replace=[True,False],value=[1,0])
df1[['Studio Apartment']]=df1['property_type']=='Studio Apartment'
df1['Studio Apartment']=df1['Studio Apartment'].replace(to_replace=[True,False],value=[1,0])
df1[['Villa']]=df1['property_type']=='Villa'
df1['Villa']=df1['Villa'].replace(to_replace=[True,False],value=[1,0])
df1[['Independent House']]=df1['property_type']=='Independent House'
df1['Independent House']=df1['Independent House'].replace(to_replace=[True,False],value=[1,0])
df1[['Penthouse']]=df1['property_type']=='Penthouse'
df1['Penthouse']=df1['Penthouse'].replace(to_replace=[True,False],value=[1,0])
print(df1.info())
df1
```

```python
df1[['BHK']]=df1['layout_type']=='BHK'
df1['BHK']=df1['BHK'].replace(to_replace=[True,False],value=[1,0])
df1[['RK']]=df1['layout_type']=='RK'
df1['RK']=df1['RK'].replace(to_replace=[True,False],value=[1,0])
print(df1.info())
df1
```
